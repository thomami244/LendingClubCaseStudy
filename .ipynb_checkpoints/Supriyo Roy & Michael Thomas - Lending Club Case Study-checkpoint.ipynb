# Removing Columns with all Null Values


```python
#Dataset Number of Nulls in Column Analysis
df.isnull().sum()
```

```python
#Deleting all columns which are having more than 99% NULL values...
To_be_deleted_columns = df.columns[df.isnull().sum()>39000]
To_be_deleted_columns
```

```python
#Checking no of Null values in these columns that need to be discarded...
df[To_be_deleted_columns].isnull().sum()
```

```python
#Deleting those columns
df=df.drop(To_be_deleted_columns,axis=1)
df.shape
```

```python
#Deleting Columns which are not required for analysis 
df.drop(['pymnt_plan','url','desc','title','zip_code','mths_since_last_delinq','pub_rec','initial_list_status','out_prncp','out_prncp_inv',
'total_pymnt','total_pymnt_inv','collection_recovery_fee','collections_12_mths_ex_med','policy_code','application_type','acc_now_delinq',
'chargeoff_within_12_mths','delinq_amnt','pub_rec_bankruptcies','tax_liens','emp_title','total_rec_int','total_rec_late_fee','recoveries',
'last_pymnt_d', 'last_pymnt_amnt', 'last_credit_pull_d'],axis=1,inplace=True)
df.head()
```

```python
df.columns
```

```python
df['loan_status'].value_counts()
```

```python
#Deleting those loan records whose status is current
df=df[df['loan_status']!='Current']
df.head()
```

```python
df['loan_status'].value_counts()
```

```python
df.shape
```

```python
#Removing % from interest rate column
df['int_rate']=df['int_rate'].apply(lambda x:x[:-2] if x[-1]=='%' else x[:-1])
df.head()
```

```python
#Checking datatypes of the remaining columns
df.dtypes
```

```python
df['int_rate']= df['int_rate'].astype(float)
df.dtypes
```

```python
df.head()
```

```python
df.dti.describe()
```

```python
#checking if the column dti contains any outliers using zscore method 
mean = df['dti'].mean()
std_dev = df['dti'].std()
df['z_scores']= np.abs((df.dti-mean)/std_dev)
df[df['z_scores']<=3]
df.head()
```

```python
df.shape
```

```python
df['issue_date']=pd.to_datetime(df.issue_d , format='%b-%y')
df.head()
```

```python
#deleting issue_d column as its formatted in DD_MM_YYYY format & a new column issue_date is created
df.drop(['issue_d'],axis=1,inplace=True)
df.head()
```

```python
#creating new columns for month & year from Issue Date Column
df['month']=df.issue_date.dt.month
df['year']=df.issue_date.dt.year
df.head()
```
